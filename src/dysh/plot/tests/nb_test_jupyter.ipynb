## Step 1: Imports

```python
from dysh import test_data_path
from dysh.plot import plot_spectrum
from dysh.fits.gbtfitsload import GBTFITSLoad
from dysh.plot.renderer import EnvironmentInfo
```

## Step 2: Check that `dysh` knows we're in Jupyter Notebook

```python
env_info = EnvironmentInfo()
env_info.info()
```

## Step 3: Load a spectrum to plot

```python
filename = test_data_path() / "TGBT21A_501_11/TGBT21A_501_11.raw.vegas.fits"
sdfits = GBTFITSLoad(filename)
psscan = sdfits.getps(scan=152, ifnum=0, plnum=0)
spectrum_to_plot = psscan.timeaverage(weights="tsys")
```

## Step 4: Test a static plot of the spectrum

```python
test_plot_static = plot_spectrum(spectrum_to_plot, interactive=False)
```

## Step 5: Test an interactive plot of the spectrum

```python
test_plot_interactive = plot_spectrum(spectrum_to_plot, interactive=True)
```
